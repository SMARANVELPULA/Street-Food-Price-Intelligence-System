```python
import pandas as pd

df = pd.read_csv("../datasets/dataset_2024.csv")

df = df.iloc[:, :-3]

print(df.head())
```

```python
for item in sorted(df["item_name"].unique()):
    print(item)
```

```python
category_summary = (
    df.groupby("category")
      .agg(
          avg_street_price=("street_price","mean"),
          avg_app_price=("app_price","mean"),
          avg_gap=("price_gap","mean"),
          avg_premium=("app_premium_percent","mean"),
          total_records=("record_id","count")
      )
      .reset_index()
)

print(category_summary)
```

```python
df.drop(
    columns=[
        "price_gap",
        "app_premium_percent"
    ],
    inplace=True
)
```

```python
print(df.columns)
```

```python
import numpy as np

restaurants = [
    "Cafe Delight",
    "Food Junction",
    "Mehfil",
    "Paradise",
    "Pista House",
    "Shah Ghouse",
    "Spice Hub",
    "Street Treats",
    "Tasty Corner",
    "Urban Bites"
]

platforms = [
    "Magicpin",
    "Swiggy",
    "Zomato"
]

ratings = [
    3.5,3.6,3.7,3.8,3.9,
    4.0,4.1,4.2,4.3,4.4,
    4.5,4.6,4.7,4.8,4.9
]

df["restaurant_name"] = np.random.choice(
    restaurants,
    size=len(df)
)

df["platform"] = np.random.choice(
    platforms,
    size=len(df)
)

df["rating"] = np.random.choice(
    ratings,
    size=len(df)
)
```

```python
df = df[
[
    "record_id",
    "date",
    "city",
    "item_name",
    "category",
    "veg_nonveg",
    "street_price",
    "app_price",
    "restaurant_name",
    "platform",
    "rating"
]
]
```
