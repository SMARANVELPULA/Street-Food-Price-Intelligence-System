```python
import pandas as pd

df = pd.read_csv("../datasets/dataset_2025.csv")

df = df.iloc[:, :-3]

print(df.head())
```

```python
print(df.isnull().sum())
```

```python
print("Duplicates:", df.duplicated().sum())
```

```python
category_summary = df.groupby("category").agg({
    "street_price": "mean",
    "app_price": "mean",
    "price_gap": "mean",
    "app_premium_percent": "mean"
}).round(2)

print(category_summary)
```

```python
print(df["category"].value_counts())
```

```python
df.groupby("category")["app_price"].mean().sort_values(ascending=False)
```

```python
category_summary = (
    df.groupby("category")
      .agg(
          avg_street_price=("street_price","mean"),
          avg_app_price=("app_price","mean"),
          avg_gap=("price_gap","mean"),
          avg_premium=("app_premium_percent","mean"),
          total_records=("record_id","count")
      )
      .reset_index()
)

print(category_summary)
```
